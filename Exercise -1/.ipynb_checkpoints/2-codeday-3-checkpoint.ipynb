```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def uni_normal(x,mean,var):
    """Univariate Distribution"""
    return ((1. / np.sqrt(2*np.pi*var)) * np.exp(-(x-mean)**2 / (2**var)))
```

```python
x = np.linspace(-4,4,num=30)
fig = plt.figure(figsize=(6,4))
plt.plot(x, uni_normal(x, mean=2, var=3), label="$\mathcal{N}(2,3)$")
plt.xlabel("$x$")
plt.ylabel("density: $p(x)$")
plt.legend(loc=1)
plt.show()
```
