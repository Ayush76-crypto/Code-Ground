```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize as spop
```

```python
def f(x):
    return x**2+15*np.cos(x)
plt.figure()
y = np.arange(-10,10,0.1)
plt.plot(y,f(y))
plt.show()
```

```python
result = spop.minimize(f, x0=0)
print(result)
print('Minmum is at: %.4f'%result.x)
plt.plot(result.x,f(result.x),'o')
```
