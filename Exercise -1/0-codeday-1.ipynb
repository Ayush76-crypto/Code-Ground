```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-np.pi,np.pi,200)
p = np.sin(2*X) # y = sin(x)
q = np.cos(2*X) # y = sin(2x)
r = np.sin(3*X) # y = sin(3x)
```

```python
plt.subplot(3,1,1)
plt.plot(X,p,color='red',lw=2)
plt.subplot(3,1,2)
plt.plot(X, q,color='green',lw=2)
plt.subplot(3,1,3)
plt.plot(X, r,color='orange',lw=2)
plt.show()
```
