```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

# FUNCTIONAL METHOD

```python
plt.plot(x,y,'r-')
```

```python
plt.plot(x,y,'g-')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Practise Graph')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r')

plt.subplot(2,1,2)
plt.plot(y,x,'g')
```

# Object oriented method

```python
fig = plt.figure()
axes = fig.add_axes()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'g')
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'g')
axes.set_xlabel('x Label')
axes.set_ylabel('Y Label')
axes.set_title('This is my title bro')
plt.show()
```
