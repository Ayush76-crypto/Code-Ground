```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=3, alpha=0.5)
# alpha --> transparency
# linewidth / lw
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',lw=4,ls='--')
# linsestyle / ls
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='magenta',lw=2,ls='-',marker='o',markersize=10)
# marker --> o,+,*,1
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='magenta',lw=3,ls='-',marker='o',markersize=15,
       markerfacecolor='yellow')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='magenta',lw=3,ls='-',marker='o',markersize=15,
       markerfacecolor='yellow',markeredgewidth=3,markeredgecolor=
       'black')
```

## Control over axis appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',lw=2,ls='--')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',lw=2,ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
